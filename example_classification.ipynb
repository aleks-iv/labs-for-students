```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
np.random.seed(512)
```

```python
data = np.loadtxt('data1.txt', delimiter=',')
```

```python
x = data[:, :2]
y = data[:, 2:]
```

```python
x_0 = x[data[:,2]==0]
x_1 = x[data[:,2]==1]
```

```python
model = LogisticRegression(C=1000, max_iter=10000)
model.fit(x, y)
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(16, 5))

plt.subplot(1, 2, 1)
plt.scatter(x_0[:,0], x_0[:,1], color='green', marker='o', label='Клас 0')
plt.scatter(x_1[:,0], x_1[:,1], color='darkred', marker='x', label='Клас 1')
plt.title('Вихідні дані')
plt.xlabel('$x1$')
plt.ylabel('$x2$')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x_0[:,0], x_0[:,1], color='green', marker='o', label='Клас 0')
plt.scatter(x_1[:,0], x_1[:,1], color='darkred', marker='x', label='Клас 1')
x1 = np.array((min(x[:,0])-100, max(x[:,1])+100))
x2 = -(model.intercept_ + model.coef_[0][0]*x1 - 0.5)/model.coef_[0][1]
plt.plot(x1, x2, color='darkorange', label='Побудована модель', scalex=False, scaley=False)
plt.title('Результат')
plt.xlabel('$x1$')
plt.ylabel('$x2$')
plt.legend()
```
