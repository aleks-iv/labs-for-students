```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
np.random.seed(512)
```

```python
clust1 = np.random.normal(5, 2, (100,2))
clust2 = np.random.normal(15, 3, (100,2))
clust3 = np.random.multivariate_normal([17,3], [[1,0],[0,1]], 30)
clust4 = np.random.multivariate_normal([2,16], [[1,0],[0,1]], 30)
X = np.concatenate((clust1, clust2, clust3, clust4))
```

```python
n_clusters = 4
model = AgglomerativeClustering(n_clusters=n_clusters)
model.fit(X)
```

```python
#centers = model.cluster_centers_
labels = model.labels_
```

```python
plt.figure(figsize=(16, 5))

plt.subplot(1, 2, 1)
plt.scatter(X[:,0], X[:,1], color='green', marker='x', label='  ')
plt.title(' ')
plt.xlabel('$x1$')
plt.ylabel('$x2$')
plt.legend()

plt.subplot(1, 2, 2)
colors = ['deepskyblue', 'darkred', 'indigo', 'orangered', 'r', 'g', 'b', 'y', 'k']
for i in range(1, n_clusters+1):
    plt.scatter(X[labels==i-1][:,0], X[labels==i-1][:,1], color=colors[i-1], marker='x', label=' ' + str(i))
#plt.scatter(centers[:,0], centers[:,1], color='k', s=100, marker='*', label='')
plt.title('')
plt.xlabel('$x1$')
plt.ylabel('$x2$')
plt.legend()
```
