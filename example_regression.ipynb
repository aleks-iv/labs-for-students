```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(512)
```

```python
x = np.linspace(-4, 21)
x = np.reshape(x, (len(x), 1))
y_clean = 2*x + 4
noise = np.random.normal(scale=6, size=y_clean.shape)
y = y_clean + noise
```

```python
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
```

```python
plt.figure(figsize=(16, 5))

plt.subplot(1, 2, 1)
plt.scatter(x, y, color='green', marker='x', label='Навчальні приклади')
plt.title('Вихідні дані')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x, y, color='green', marker='x', label='Навчальні приклади')
plt.plot(x, y_pred, color='darkorange', label='Побудована модель')
plt.title('Результат')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
```
