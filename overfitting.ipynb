```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
def model_train(x, y):
    model = LinearRegression()
    model.fit(x, y)
    return model.predict(x)
```

```python
np.random.seed(1024)
```

```python
x = np.arange(-0.7, 3.8, 0.22)
x = np.reshape(x, (len(x), 1))
y_real = np.sin(x)
noise = np.random.normal(scale=0.4, size=y_real.shape)
y = y_real + noise
```

```python
x_underfit = x
```

```python
x_robust = np.concatenate((x, np.power(x, 2)), axis=1)
```

```python
x_overfit = np.concatenate((x, np.power(x, 2), np.power(x, 3), np.power(x, 4),
                           np.power(x, 5), np.power(x, 6), np.power(x, 7),
                           np.power(x, 8), np.power(x, 9), np.power(x, 10),
                           np.power(x, 11), np.power(x, 12), np.power(x, 13),
                           np.power(x, 14), np.power(x, 15), np.power(x, 16),
                           np.power(x, 17), np.power(x, 18), np.power(x, 19)), axis=1)
```

```python
y_pred_underfit = model_train(x_underfit, y)
y_pred_robust = model_train(x_robust, y)
y_pred_overfit = model_train(x_overfit, y)
```

```python
plt.figure(figsize=(20, 12))

plt.subplot(2, 2, 1)
plt.scatter(x, y, lw=5, color='#edc951', label='Данні із шумом')
plt.plot(x, y_real, color='#eb6841', label='Істинна залежність')
plt.legend()
plt.title('Вихідні дані')

plt.subplot(2, 2, 2)
plt.scatter(x, y, lw=5, color='#edc951', label='Данні із шумом')
plt.plot(x, y_pred_underfit, lw=3, color='#00a0b0', label='Модель')
plt.plot(x, y_real, color='#eb6841', label='Істинна залежність')
plt.legend()
plt.title('Модель степеня 1')

plt.subplot(2, 2, 3)
plt.scatter(x, y, lw=5, color='#edc951', label='Данні із шумом')
plt.plot(x, y_pred_robust, lw=3, color='#00a0b0', label='Модель')
plt.plot(x, y_real, color='#eb6841', label='Істинна залежність')
plt.legend()
plt.title('Модель степеня 2')

plt.subplot(2, 2, 4)
plt.scatter(x, y, lw=5, color='#edc951', label='Данні із шумом')
plt.plot(x, y_pred_overfit, lw=3, color='#00a0b0', label='Модель')
plt.plot(x, y_real, color='#eb6841', label='Істинна залежність')
plt.legend()
plt.title('Модель степеня 19')
```
