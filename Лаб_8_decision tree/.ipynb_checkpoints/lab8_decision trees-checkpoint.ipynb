```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import scale
from sklearn.datasets import load_iris, load_wine, load_breast_cancer
%matplotlib inline
```

```python
data = load_iris()
```

```python
n_classes = len(data.target_names)
plot_step = 0.02
plot_colors = ['darkblue', 'darkorange', 'g']
```

```python
# Оберемо дві ознаки (для забезпечення можливості побудови графіка)
f1 = 3
f2 = 2
X = data.data[:,[f1, f2]]
X = scale(X, axis=0)
y = data.target
X_names = [data.feature_names[f1], data.feature_names[f2]]
y_names = data.target_names
```

```python
plt.figure(figsize=(7,5))
for i, color in zip(range(n_classes), plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, label=y_names[i], cmap=plt.cm.RdYlBu, edgecolor='k')
plt.xlabel(X_names[0])
plt.ylabel(X_names[1])
plt.legend(y_names)
```

```python
tree = DecisionTreeClassifier(random_state=57)
tree.fit(X, y)
```

```python
plt.figure(figsize=(20,7))

plt.subplot(1, 2, 1)
x_min, x_max = X[:, 0].min()-0.3, X[:, 0].max()+0.3
y_min, y_max = X[:, 1].min()-0.3, X[:, 1].max()+0.3
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))

Z = tree.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.3)

plt.xlabel(X_names[0])
plt.ylabel(X_names[1])

# Plot the training points
for i, color in zip(range(n_classes), plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, label=y_names[i], cmap=plt.cm.RdYlBu, edgecolor='k')

plt.suptitle("Decision surface of a decision tree")
plt.legend()

plt.subplot(1, 2, 2)

plot_tree(tree, feature_names=X_names, class_names=y_names, filled=True)
```
